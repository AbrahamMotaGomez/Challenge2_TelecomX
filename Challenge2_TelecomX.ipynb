```python
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

df = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)
display(df.head())
```

```python
lista_est=[
 'Churn',
 'Partner',
 'Dependents',
 'PhoneService',
 'MultipleLines',
 'OnlineSecurity',
 'OnlineBackup',
 'DeviceProtection',
 'TechSupport',
 'StreamingTV',
 'StreamingMovies', 'PaperlessBilling']

df[lista_est]=df[lista_est].replace("Yes","1").astype(str)
```

```python
fig, ax=plt.subplots(4,2,figsize=(10,10))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Efectos del tipo de clientes para concelar servicio")

ax[0,0].bar(inactivos["gender"].value_counts().index,inactivos["gender"].value_counts())
ax[0,0].set_title("Inactivos - Efecto de GÃ©nero")

# Generar marcadores tipo texto para valores numericos
marcadores_vejez = {0: 'No viejo', 1: 'Viejo'}
cuenta_vejez = inactivos["SeniorCitizen"].value_counts().rename(index=marcadores_vejez)
ax[0,1].bar(cuenta_vejez.index, cuenta_vejez.values)
ax[0,1].set_title("Inactivos - Efecto de Vejez")

ax[1,0].bar(inactivos["Contract"].value_counts().index,inactivos["Contract"].value_counts())
ax[1,0].set_title("Inactivos - Efecto por tiempo de contrato")

cuenta_civil={0:"Soltero",1:"Casado"}
cuenta_civil=inactivos["Partner"].value_counts().rename(index=cuenta_civil)
ax[1,1].bar(cuenta_civil.index,cuenta_civil.values)
ax[1,1].set_title("Inactivos - Efecto por estado civil")

ax[2,0].bar(Activos["gender"].value_counts().index,Activos["gender"].value_counts(),color="green")
ax[2,0].set_title("Activos - Efecto de GÃ©nero")

# Generar marcadores tipo texto para valores numericos
marcadores_vejez_2 = {0: 'No viejo', 1: 'Viejo'}
cuenta_vejez_2 = Activos["SeniorCitizen"].value_counts().rename(index=marcadores_vejez_2)
ax[2,1].bar(cuenta_vejez_2.index, cuenta_vejez_2.values,color="green")
ax[2,1].set_title("Activos - Efecto de Vejez")

ax[3,0].bar(Activos["Contract"].value_counts().index,Activos["Contract"].value_counts(),color="green")
ax[3,0].set_title("Activos - Efecto por tiempo de contrato")

cuenta_civil_2={0:"Soltero",1:"Casado"}
cuenta_civil_2=Activos["Partner"].value_counts().rename(index=cuenta_civil_2)
ax[3,1].bar(cuenta_civil_2.index,cuenta_civil_2.values,color="green")
ax[3,1].set_title("Activos - Efecto por estado civil")

for ax in ax.flat:
  ax.grid(True)

plt.show()
```

```python
fig, ax=plt.subplots(2,2,figsize=(10,10))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Efectos del tiempo y cobro mensual de contrato")

ax[0,0].barh(inactivos["tenure"].value_counts().index,inactivos["tenure"].value_counts())
ax[0,0].set_title("Inactivos - Tiempo de contrato en meses")

ax[0,1].hist(inactivos['Charges.Total'].dropna(), bins=30, alpha=0.5, label='Activos', color='green')
ax[0,1].set_title("Inactivos - Efecto por cobro total")

ax[1,0].barh(Activos["tenure"].value_counts().index,Activos["tenure"].value_counts(),color="purple")
ax[1,0].set_title("Activos - Tiempo de contrato en meses")

ax[1,1].hist(Activos['Charges.Total'].dropna(), bins=30, alpha=0.5, label='Inactivos', color='red')
ax[1,1].set_title("Activos - Efecto por cobro total")


for ax in ax.flat:
  ax.grid(True)

plt.show()
```
